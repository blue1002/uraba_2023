```python
dir='/content/drive/MyDrive/working_Angie_data_base_2023/DATA'
```

```python
all_files = glob.glob(dir + "/*.csv") #*/.csv todos y solo archivos csv
```

```python
final_df = pd.DataFrame()
```

```python
for filename in all_files: 
  nombre=os.path.basename(filename)

  df= pd.read_csv(filename, sep=',',header=3,parse_dates=['FORMATTED DATE_TIME'])
  #final_df=final_df.append(df)#Adiciona el dataframe leido al dataframe final
  separado=nombre.split('_')#utlizo split PARA QUE ME SEPARE POR _,El método split(sep=None, maxsplit=-1) en Python devuelve una lista de palabras o tokens usando sep como cadena de separación. 
    #split Básicamente, se utiliza para dividir o separar un string en partes.
  st_name=separado[0]#Obtiene la estacion 
  cod_sensor=separado[1] #Obtiene el codigo del sensor 
    #Copia en una columna del dataframe el nombre de la estacion en otra la version y por ultimo en una tercera la ubicacion de la estacion
  df['st_name']=st_name
  df['cod_sensor']=cod_sensor
  final_df=final_df.append(df)#Adiciona el dataframe leido al dataframe final
  print(df)
  print('File: '+filename + ' Read (rows,cols)', df.shape)
print(final_df.head())
```
